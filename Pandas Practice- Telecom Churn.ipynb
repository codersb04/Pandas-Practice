## Indexing and retrieving the data

## Applying Functions to cells, COlumns and Rows

```python
df.apply(np.max)
```

```python
df[df["State"].apply(lambda state: state[0] == "W")]
```

```python
# Replacing the values of column using map
d = {"No": False, "Yes":True}
df["International plan"]=df["International plan"].map(d)
df["International plan"].head()
```

```python
# Replace using replace function
df = df.replace({"Voice mail plan":d})
df["Voice mail plan"].head()
```

## Grouping

```python
column_to_show = ["Total day minutes", "Total eve minutes",
                  "Total night minutes"]
df.groupby(["Churn"])[column_to_show].describe()
```

```python
df.groupby(["Churn"])[column_to_show].agg([np.mean, np.std, np.min, np.max])
```

## Summary Table

```python
# Data distribution with respect to 2 columns
pd.crosstab(df["Churn"],df["International plan"])
```

```python
pd.crosstab(df["Churn"],df["Voice mail plan"],normalize=True)
```

####  Summarizing the dataset using Pivot Table
Parameters for pivot tables are:
value: list of variables to calculate statistics
index: a list of variable to group by
aggfunc: What statistics we need to calculate for the group eg mean,sum, min,max etc

```python
df.pivot_table(
["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
aggfunc="mean")
```

## DataFrame transformations

```python
# Adding new column to dataframe using the existing columns
# Like here we can add total number of call from user
total_calls = (df["Total day calls"]+df["Total eve calls"]
              +df["Total night calls"]+df["Total intl calls"])

df.insert(loc = len(df.columns), column = "Total calls",
          value = total_calls)
# loc parameter is the number of column after which we need to insert the new column
# here by specifing len(df.columns), new column will be added at the end to the dataframe
```

```python
df.head()
```

```python
# new column can also be added without using insert function
df["Total charges"] = (df["Total day charge"]+df["Total eve charge"]
                      +df["Total intl charge"]+df["Total night charge"])

df.head()
```

```python
# Deleting a column from dataframe
df.drop(["Total charges", "Total calls"], axis = 1, inplace = True)
```

```python
# Deleting the rows from the dataframe
df.drop([1,2]).head()
```
