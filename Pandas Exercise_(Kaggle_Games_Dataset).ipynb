## Data Visualization

```python
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
%matplot inline
```

```python
game.dropna(inplace=True)
```

```python
game.isnull().sum()
```

```python
sns.distplot(game['average_rating'])
plt.show()
```

-- Inference: lots of 0 value

```python
sns.jointplot(x ='minage', y ='average_rating', data =game)
plt.show()
```

```python
sns.pairplot(game[['playingtime','minage','average_rating']])
```

```python
# compare type of game and playtime using a stripplot
sns.stripplot(x ='type', y='playingtime', data=game, jitter=True)
```

```python
# Analyse the linear trend between playing time(less than 500 mins) and average_rating received for the same
sns.regplot(x='playingtime', y ='average_rating', data=game[game['playingtime']<500])
```
